## IMPLEMENTATION OF HUFFMAN CODING 

**Name:** Rithika L  
**Reg No:** 212224230231  
**Slot:** T1-H15


## 1: Set the input string

```python
input_string = "RITHIKA L"
```

## 2: Calculate frequency of each character in the input string

```python
frequency = {}
for char in input_string:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
```

## 3: Create tree nodes

```python
nodes = [[char, freq] for char, freq in frequency.items()]
```

## 4: Main function to implement Huffman coding

```python
while len(nodes) > 1:
    nodes = sorted(nodes, key=lambda x: x[1])
    left = nodes.pop(0)
    right = nodes.pop(0)
    new_node = [[left, right], left[1] + right[1]]
    nodes.append(new_node)
huffman_tree = nodes[0]
```

## 5: Generate Huffman codes

```python
huffman_codes = {}

def generate_codes(tree, code=""):
    if isinstance(tree[0], str):  # If it's a leaf node
        huffman_codes[tree[0]] = code
    else:  
        generate_codes(tree[0][0], code + "0")
        generate_codes(tree[0][1], code + "1")

generate_codes(huffman_tree)
```

## 6: Print the characters and their Huffman codes

```python
print("Character | Huffman Code")
print("-------------------------")
for char, code in huffman_codes.items():
    print(f"    {char}    |    {code}")
```
